```python
# Step 1: Install the necessary libraries
!pip install -q transformers gradio

# Step 2: Import libraries
from transformers import pipeline
import gradio as gr

# Step 3: Load the sentiment analysis pipeline
# This downloads a pre-trained model automatically
classifier = pipeline("sentiment-analysis")

# Step 4: Define a simple function for the app to use
def analyze_text(text):
    result = classifier(text)[0]
    return f"Label: {result['label']}, Score: {round(result['score'], 4)}"

# Step 5: Create the web interface
demo = gr.Interface(
    fn=analyze_text,
    inputs=gr.Textbox(lines=2, placeholder="Type something..."),
    outputs="text",
    title="My First AI Sentiment App",
    description="Type a sentence below to see if it's Positive or Negative!"
)

# Step 6: Launch the app
demo.launch(share=True)
```

```python
!pip install -q transformers gradio torch

from transformers import pipeline
import gradio as gr

# text-generation is available in your environment
gen = pipeline("text-generation", model="google/flan-t5-base")

def summarize(text):
    text = text.strip()
    if not text:
        return "Please paste some text."
    prompt = f"Summarize this in 5-7 bullet points:\n\n{text}"
    out = gen(prompt, max_new_tokens=160, do_sample=False)[0]["generated_text"]
    # flan sometimes echoes prompt; keep only after prompt if needed
    return out.replace(prompt, "").strip()

demo = gr.Interface(
    fn=summarize,
    inputs=gr.Textbox(lines=10, placeholder="Paste long text here..."),
    outputs="text",
    title="Summarizer (HF via Text-Generation)",
    description="Uses a text-generation model to summarize."
)

demo.launch(share=True)
```

```python
!pip install -q transformers gradio pillow torch

from transformers import pipeline
import gradio as gr
from PIL import Image

img_clf = pipeline("image-classification", model="google/vit-base-patch16-224")

def classify_image(img: Image.Image):
    if img is None:
        return "Upload an image first."
    preds = img_clf(img)
    top5 = preds[:5]
    return "\n".join([f"{i+1}. {p['label']}  (score: {p['score']:.4f})" for i, p in enumerate(top5)])

gr.Interface(
    fn=classify_image,
    inputs=gr.Image(type="pil"),
    outputs="text",
    title="Image Classification App",
    description="Upload an image and it predicts what it is."
).launch(share=True)
```
